```python
# Carregar datasets
data = pd.read_excel('heart.xlsx')
data.head()
```

```python
# 1. Efetue a criação de um gráfico de barras (barplot ou countplot) para a comparação da distribuição de sex(gênero).
sns.countplot(x='sex', data=data)
plt.title('Distribuição por Sexo')
plt.show()
```

```python
# 2. Efetua a criação de um histograma para visualizar a distribuição da idade dos pacientes separados por sex (gênero), incluindo a plotagem de média e mediana.
sns.histplot(data=data, x='age', hue='sex', kde=True)
plt.axvline(data['age'].mean(), color='k', linestyle='--')
plt.axvline(data['age'].median(), color='r', linestyle='-')
plt.title('Distribuição da Idade por Sexo')
plt.legend(['Média', 'Mediana'])
plt.show()
```

```python
# 3. Efetue a criação de um boxplot para análise das variáveis de idade, cholestoral (Chol), pressão sanguínea (trestbps), e máxima frequência cardíaca. Identifique as variáveis por
sns.boxplot(data=data[['age', 'chol', 'trestbps', 'thalach']])
plt.title('Boxplot das Variáveis de idade')
plt.show()
```

```python
# 4. Efetue a criação de um gráfico de pizza, para apresentar a distribuição dos tipos de dor no peito (cp), identificando-os no gráfico.
plt.pie(data['cp'].value_counts(), labels=data['cp'].unique(), autopct='%1.1f%%')
plt.title('Distribuição dos Tipos de Dor no Peito')
plt.show()
```

```python
# 5. Efetue a criação de um gráfico de linha para apresentar a frequência cardíaca máxima dos pacientes (talach).
sns.lineplot(data=data, x=data.index, y='thalach')
plt.title('Frequência Cardíaca Máxima')
plt.xlabel('Índice')
plt.ylabel('Frequência Cardíaca Máxima')
plt.show()
```

```python
# 6. Criar um grafico de barras para avaliar a distribuição de tipos de dor no peito (cp), separados pela variável alvo (identifica a existência ou não de doença cardíaca).
sns.countplot(x='cp', hue='target', data=data)
plt.title('Distribuição de Tipos de Dor no Peito por Target')
plt.show()
```

```python
# 7. Efetue a criação de um pair plot para avaliar a correlação e distribuição das variáveis idade, colesterol (Chol), pressão sanguínea (trestbps), frequência cardíaca (talach), depressão do segmento ST (oldpeake) e a variável alvo (target).
sns.pairplot(data[['age', 'chol', 'trestbps', 'thalach', 'oldpeak', 'target']])
plt.show()
```

#Interpretação das visualizações criadas

Distribuição de sexo: Se a maioria dos pacientes for do sexo masculino, isso pode indicar um viés na coleta de dados ou uma característica específica da população estudada.
Histograma de idade: A distribuição da idade pode indicar se a doença cardíaca afeta mais pessoas de determinada faixa etária. A comparação entre os sexos pode revelar diferenças.
Boxplot: Os boxplots ajudam a identificar outliers, a variabilidade dos dados e a distribuição dos valores.
Gráfico de pizza: A distribuição dos tipos de dor no peito pode indicar a prevalência de diferentes tipos de dor entre os pacientes.
Gráfico de linha: O gráfico de linha da frequência cardíaca máxima pode mostrar se há alguma tendência ao longo do tempo (se os dados estiverem ordenados cronologicamente).
Gráfico de pares: O pair plot permite visualizar a relação entre todas as variáveis numéricas e a variável alvo. A presença de correlações pode indicar quais variáveis são mais importantes para prever a doença cardíaca.

Ao analisar os gráficos, é possivel identificar:

- Fatores de risco
- Diferenças entre os sexos
- Grupos de pacientes
